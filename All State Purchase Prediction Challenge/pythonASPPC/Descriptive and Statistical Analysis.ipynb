The dataset contains information above the behavior of customers who purchased an insurance policy. It contains customer data like 'car age', 'car value', 'risk factor' and also information about what insurance policies 'they looked at before making a purchase, these are represented by 7 columns A to G each of these options has 2,3 or 4 options. Based on the options available each customer is picking an insurnace policy out of 2304 options available. Each record in the dataset is a viewing point, meaning that it is a configuration of an insurance policy that the client looked at before making a purchase. The final record for each customer is what the client actually purchased. A record type of value '0' is a record that the customer looked at and didn't make a purchase, while a record type of value '1' is a record type were the customer made a purchase. The customer can view the policy they purchase multiple times before actually deciding if that is the correct choice. For example if a customer has 10 records in the system, 9 of them will have the value of zero and the last one will have the value of 1 which will indicate what the client bought. What makes this problem complicated is the fact that we don't have any information about how the cost is being calculated or what are the options the client is picking to see if we can correlate them back to the charactistics. Below is some statistical analysis on the dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import pylab as pl
%matplotlib inline

data = pd.read_csv("../data/train.csv")
```

```python

print "number of records: ", len(data) 
print "number of unique customer_IDs: ", len(data['customer_ID'].unique())
```

```python
print "columns: ", data.columns
```

Number of Nulls in each column:
    

```python
#count nulls in each columns
for column in data.columns:
    print column, ": ", len(data[pd.isnull(data[column])])
```

```python
data.groupby(data['shopping_pt']).size().plot(kind='bar'\
                                              ,title='Number of shopping pt before purchase')
```

```python
s = np.zeros(11)
for i in range(2,13):
    x_shopping_pt = data[data['shopping_pt'] > i]
    x_m1_shopping_pt = data[data['shopping_pt'] > i - 1]
    s[i - 2] = (len(x_m1_shopping_pt['customer_ID'].unique()))\
    - (len(x_shopping_pt['customer_ID'].unique()))

spd = pd.Series(s)
spd.plot(kind='bar',title="shopping points before purchase customer count")
```

```python
data.groupby(data['state'])['cost'].mean().plot(kind='bar'\
                                                ,title='Average cost per State',figsize=(20,5))
```

```python
data.groupby(data['state'])['cost'].std().plot(kind='bar'\
                                               ,title='Cost standard deviation per state',figsize=(20,5))
```

```python
h = sorted(np.asarray(data['cost']))
fit = stats.norm.pdf(h, np.mean(h), np.std(h))  
pl.plot(h,fit,'-o')
pl.hist(h,normed=True) 
pl.title('distribution of cost')
pl.show()                  
```
