```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import pylab as pl
%matplotlib inline

data = pd.read_csv("../data/train.csv")
```

```python

print "number of records: ", len(data)  
print "number of unique customer_IDs: ", len(data['customer_ID'].unique())
```

```python
print "columns: ", data.columns
```

Number of Nulls in each column:
    

```python
#count nulls in each columns
for column in data.columns:
    print column, ": ", len(data[pd.isnull(data[column])])
```

```python
data.groupby(data['shopping_pt']).size().plot(kind='bar',title='Number of shopping pt before purchase')
```

```python
data.groupby(data['state'])['cost'].mean().plot(kind='bar',title='Average cost per State',figsize=(20,5))
```

```python
data.groupby(data['state'])['cost'].std().plot(kind='bar',title='Cost standard deviation per state',figsize=(20,5))
```

```python
h = sorted(np.asarray(data['cost']))

fit = stats.norm.pdf(h, np.mean(h), np.std(h))  
pl.plot(h,fit,'-o')
pl.hist(h,normed=True) 
pl.title('distribution of cost')
pl.show()                  
```
